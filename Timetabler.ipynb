```python
import pandas as pd
import numpy as np
import random as r
from Rating import Subject
import Rating as rt

table1 = pd.read_csv("TimetablesB.csv", index_col= 0)
```

```python
def add_subs(tt,subjects):
    redo= 0
    dates = ['M', 'TU', 'W', 'TH', 'F']
    for sub in subjects:
        free = True
        while free == True:
            dur = sub.units
            date = r.choice(['M', 'TU', 'W', 'TH', 'F'])
            start = r.randrange(0,8)
            if tt.loc[start + dur-1,date] == '-' and tt.loc[start,date] =='-':
                while dur >= 0:
                    
                    if dur > 0:
                        print("placing:", sub.code)
                        tt.loc[start,date] = sub.code
                    else:
                        free =False
                    dur -= 1
                    start += 1
                    
            else:
                print("there's something there, couldn't place", sub.code)
                #tt = tt.replace(sub.code,'-')
    tt = rt.timetable_to_values(tt)
    tt = rt.de_nan(tt)
    tt = rt.normnew(table1,tt)
    tt = tt.drop(['C/semester', 'Combined Score', 'Score based Rating'], axis = 1)
    tt = tt.reindex(columns = ['total sessions', 'sess/popday', 'Number of Blocks', 'Block/pday',
       'Avergae Block size', 'Norm Block', 'Early Sessions', 'Late/day',
       'Avg Break'])  
    redo = rt.feedback(tt)

    return redo
  
```
